```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
# Đọc dữ liệu
df = pd.read_csv("ML_DATA_TEST.csv")
```

```python
df
```

```python
# Xóa các bản ghi có trạng thái không cần thiết
data = df[~df['trangthai'].isin([0, 4, 5,8])]

# Kiểm tra danh sách các cột
print("Cột trong DataFrame:", data.columns)


# Điền giá trị NA bằng 0
data.fillna(0, inplace=True)

# Sắp xếp dữ liệu
data.sort_values(by=['namhoc', 'hocky'], ascending=True, inplace=True)
```

```python
label_encoder = LabelEncoder() #encode cột dữ liệu phân loại

data['mssv'] = label_encoder.fit_transform(data['mssv'])
data = pd.get_dummies(data, columns=[' khoa', ' hedt'], drop_first=True)
data['mamh'] = label_encoder.fit_transform(data['mamh'])
train_data = data.loc[
    (data['namhoc'] < 2021) | ((data['namhoc'] == 2021) & (data['hocky'] == 1))
]

# Tách đặc trưng và mục tiêu
X = train_data.drop(['diem_hp'], axis=1)  # Đặc trưng
y = train_data['diem_hp']
```

```python
# Chia dữ liệu thành tập huấn luyện (80%) và tập còn lại (20%)
train_data, temp_data = train_test_split(data, test_size=0.2, random_state=42)

# Chia tập còn lại thành tập validate (10%) và tập test (10%)
validate_data, test_data = train_test_split(temp_data, test_size=0.5, random_state=42)
```

```python
# Tách đặc trưng và nhãn cho tập huấn luyện
X_train = train_data.drop(['diem_hp'], axis=1)
y_train = train_data['diem_hp']

# Tách đặc trưng và nhãn cho tập validate
X_validate = validate_data.drop(['diem_hp'], axis=1)
y_validate = validate_data['diem_hp']

# Tách đặc trưng và nhãn cho tập test
X_test = test_data.drop(['diem_hp'], axis=1)
y_test = test_data['diem_hp']

# Chuyển đổi sang mảng NumPy
X_train_np = X_train.to_numpy()
y_train_np = y_train.to_numpy()
X_validate_np = X_validate.to_numpy()
y_validate_np = y_validate.to_numpy()
X_test_np = X_test.to_numpy()
y_test_np = y_test.to_numpy()
# In ra các giá trị của X_train để kiểm tra
print(X_train)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dropout

X_train_np = np.array(X_train).astype(np.float32)
y_train_np = np.array(y_train).astype(np.float32)

X_validate_np = np.array(X_validate).astype(np.float32)
y_validate_np = np.array(y_validate).astype(np.float32)

X_test_np = np.array(X_test).astype(np.float32)
y_test_np = np.array(y_test).astype(np.float32)


scaler = StandardScaler()
X_train_np = scaler.fit_transform(X_train_np)
X_validate_np = scaler.transform(X_validate_np)
X_test_np = scaler.transform(X_test_np)

# Xây dựng mô hình Neural Network
model = Sequential()
model.add(Dense(128, input_dim=X_train_np.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))  # Lớp đầu ra
model.add(Dropout(0.2))  # Thêm lớp dropout
# Biên dịch mô hình
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])

# Huấn luyện mô hình
model.fit(X_train_np, y_train_np, epochs=50, batch_size=128, validation_data=(X_validate_np, y_validate_np))

# Đánh giá mô hình trên tập kiểm tra
test_loss, test_mae = model.evaluate(X_test_np, y_test_np)
print(f'Test loss: {test_loss}, Test MAE: {test_mae}')

# Dự đoán trên tập kiểm tra
predictions = model.predict(X_test_np)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Tính MSE, RMSE, MAE và R-squared
mse = mean_squared_error(y_test_np, predictions)
rmse = np.sqrt(mse)  # Tính RMSE từ MSE
mae = mean_absolute_error(y_test_np, predictions)
r_squared = r2_score(y_test_np, predictions)  # Tính R-squared

print(f'Test MSE: {mse}, Test RMSE: {rmse}, Test MAE: {mae}, Test R-squared: {r_squared}')
```
